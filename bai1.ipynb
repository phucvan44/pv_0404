```python
import numpy as np
from math import sqrt
```

```python
def gen_array(lim, sizeX, sizeY):
    return np.random.randint(lim, size = (sizeX, sizeY))
```

A)

```python
A = gen_array(10, 3, 4)
B = gen_array(10, 4, 4)
print("A: ", A, "\n", '-'*20)
print("B: ", B, "\n", '-'*20)
```

B)


```python
A = gen_array(10, 4, 4)
B = ~(A%2) + 2
print("A: ", A, "\n", '-'*20)
print("B: ", B, "\n", '-'*20)
```

C)

```python
def prime(n):
    if (n%2 == 0 and n > 2) or (n < 2):
        return False
    return all(n%i for i in range(3, int(sqrt(n) + 1), 2))
```

```python
A = gen_array(10, 5, 5)
print("A: ", A, "\n", '-'*20)
A = np.fliplr(A)
func = np.vectorize(prime)
A_temp = func(A.diagonal())
count_prime = len(np.extract(A_temp, A))
print("A flip left to right: ", A, "\n", '-'*20)
print(count_prime)
```
